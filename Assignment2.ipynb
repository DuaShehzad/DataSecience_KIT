```python
import pandas as pd
```

```python
import numpy as np
```

```python
movies_data=pd.read_excel("/content/movies.xlsx")
```

```python
movies_data.head(5)
```

```python
movies_data=movies_data.rename(columns=({'MOVIES':'RENAME MOVIES'}))
```

```python
movies_data.info()
```

```python
movies_data['YEAR']=movies_data['YEAR'].str.replace('('," ").str.replace(')'," ").str.replace(r'[()\u2013\u2014-]'," ")
```

```python
movies_data['YEAR']=movies_data['YEAR'].str.strip()
```

```python
movies_data.head(5)
```

```python
movies_data['STARS']=movies_data['STARS'].str.replace("\n"," ")
```

```python

split_data=movies_data['STARS'].str.strip().str.split('|', 1,expand=True)
```

```python
if len(split_data.columns)==2 :
  movies_data['Director']=split_data[0]
  movies_data['Star']=split_data[1]
else:
  movies_data['Director']=""
  movies_data['Star']=movies_data['STARS']
```

```python
movies_data.head(5)
```

```python
movies_data['GENRE']=movies_data['GENRE'].str.strip('\n')
```

```python
movies_data['ONE-LINE']=movies_data['ONE-LINE'].str.strip('\n')
```

```python
movies_data.isnull().sum()
```

```python
movies_data[['RATING','VOTES','RunTime']]=movies_data[['RATING','VOTES','RunTime']].fillna(0.0)
```

```python
movies_data[['YEAR','GENRE','Gross','Star']]=movies_data[['YEAR','GENRE','Gross','Star']].fillna('0')
```
